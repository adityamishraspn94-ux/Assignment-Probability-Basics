## **Question 1:**

A die is rolled. What is the probability of getting:
**(a)** an even number
**(b)** a number greater than 4

## **Solution:**

A standard die has **6 outcomes**: {1, 2, 3, 4, 5, 6}
Total outcomes = **6**

###**(a) Probability of getting an even number**

Even numbers on a die = **2, 4, 6**
Number of favorable outcomes = **3**

[
P(\text{even number}) = \frac{3}{6} = \frac{1}{2}
]

### **(b) Probability of getting a number greater than 4**

Numbers greater than 4 = **5, 6**
Number of favorable outcomes = **2**

[
P(\text{number > 4}) = \frac{2}{6} = \frac{1}{3}
]

## **Final Answers:**

* **(a)** Probability of getting an even number = **1/2**
* **(b)** Probability of getting a number greater than 4 = **1/3**

## **Question 2:**

In a class of 50 students:

* 20 like Mathematics (M)
* 15 like Science (S)
* 5 like both subjects

What is the probability that a student chosen at random **likes Mathematics or Science**?
## **Solution:**

We use the formula for **union of two sets**:

[
P(M \cup S) = \frac{n(M) + n(S) - n(M \cap S)}{ \text{Total students} }
]

Where:

* ( n(M) = 20 )
* ( n(S) = 15 )
* ( n(M \cap S) = 5 )
* Total students = 50
### **Step 1: Find number of students who like Mathematics OR Science**

[
n(M \cup S) = 20 + 15 - 5 = 30
]

### **Step 2: Probability**

[
P(M \cup S) = \frac{30}{50} = \frac{3}{5}
]

## **Final Answer:**

**Probability that a randomly selected student likes Mathematics or Science = 3/5**

## **Question 3:**

A bag has **3 red** and **2 blue** balls.
If one ball is drawn randomly **and is red**, what is the probability that the **next ball is also red** (without replacement)?

## **Solution:**

### **Step 1: Original balls in the bag**

* Red = 3
* Blue = 2
* Total = 5
### **Step 2: First ball drawn is red**

Since the first ball is **already known to be red**, we remove one red ball.

Remaining balls:

* Red left = **3 − 1 = 2**
* Blue = 2
* Total remaining = **4**

### **Step 3: Probability next ball is red**

[
P(\text{second ball is red}) = \frac{\text{Remaining red balls}}{\text{Remaining total balls}}
]

[
P = \frac{2}{4} = \frac{1}{2}
]

## **Final Answer:**

**Probability that the next ball is also red = 1/2**

## **Question 4:**

The population of a school is divided into **60% boys** and **40% girls**.
If you want **equal representation of both genders** in the sample, which method should you use:
**Simple Random Sampling or Stratified Sampling? Why?
## **Answer: Stratified Sampling**

### **Explanation:**

* **Simple Random Sampling** selects students purely by chance.
  Because the population has **60% boys**, the sample will also roughly have **more boys than girls**.
  This means **equal representation is not guaranteed**.

* **Stratified Sampling** divides the population into **strata (groups)** — in this case:

  * One stratum = Boys
  * Second stratum = Girls

  Then you take **equal samples from each group**.

### Therefore:

* **Stratified Sampling ensures equal representation of boys and girls**
* It gives a **balanced, fair, and accurate sample**, especially when groups have different proportions.
## **Final Answer:**

**Use Stratified Sampling because it divides the population into boys and girls and allows you to select equal numbers from each group, ensuring proper representation.**

## **Question 5:**

The average height of **1000 students = 160 cm**.
A sample of **100 students shows an average height = 158 cm**.

Find the **sampling error**.

## **Solution:**

### **Formula for Sampling Error:**

[
\text{Sampling Error} = \text{Population Mean} - \text{Sample Mean}
]

Given:

* Population Mean ( \mu = 160 ) cm
* Sample Mean ( \bar{X} = 158 ) cm

[
\text{Sampling Error} = 160 - 158 = 2 \text{ cm}
]

## **Final Answer:**

**Sampling Error = 2 cm**

## **Question 6:**

The population mean salary is ₹50,000 with ( \sigma = ₹5,000 ).
If we take a sample of **100 employees**, what is the **Standard Error of the Mean (SEM)**?
## **Solution:**

### **Formula:**

[
SEM = \frac{\sigma}{\sqrt{n}}
]

Given:

* Population standard deviation ( \sigma = 5000 )
* Sample size ( n = 100 )

### **Calculate SEM**

[
SEM = \frac{5000}{\sqrt{100}} = \frac{5000}{10} = 500
]

## **Final Answer:**

**Standard Error of the Mean (SEM) = ₹500**

## **Question 7:**

In a group of 100 students:

* 40 like Cricket (C)
* 30 like Football (F)
* 10 like both Cricket and Football

Find the **probability that a student likes at least one sport**.

## **Solution:**

To find students who like **at least one sport (C ∪ F)**, we use:

[
n(C \cup F) = n(C) + n(F) - n(C \cap F)
]

Substitute values:

[
n(C \cup F) = 40 + 30 - 10 = 60
]

So, **60 students** like at least one sport.

### **Probability:**

[
P(C \cup F) = \frac{60}{100} = \frac{3}{5}
]

## **Final Answer:**

**Probability that a student likes at least one sport = 3/5**

## **Question 8:**

From a deck of **52 cards**, two cards are drawn **without replacement**.
What is the **probability that both are Aces**?

## **Solution:**

A standard deck has **4 Aces**.

### **Step 1: Probability the first card is an Ace**

[
P(\text{1st Ace}) = \frac{4}{52}
]

### **Step 2: Probability the second card is also an Ace (after one Ace is removed)**

Remaining cards = 51
Remaining Aces = 3

[
P(\text{2nd Ace} \mid \text{1st Ace}) = \frac{3}{51}
]

### **Step 3: Total probability**

[
P(\text{Both Aces}) = \frac{4}{52} \times \frac{3}{51}
]

Simplify:

[
= \frac{4}{52} \cdot \frac{3}{51}
= \frac{1}{13} \cdot \frac{3}{51}
= \frac{3}{663}
= \frac{1}{221}
]

## **Final Answer:**

**Probability that both cards drawn are Aces = 1/221**

## **Question 9:**

A factory produces bulbs with a **2% defective rate**.
If **5 bulbs** are chosen at random, what is the probability that **all are non-defective**?

## **Solution:**

Defective rate = **2%**
So, **non-defective rate = 100% – 2% = 98% = 0.98**

We want all 5 bulbs to be **non-defective**.

Because each selection is independent:

[
P(\text{all 5 non-defective}) = (0.98)^5
]

### **Calculate:**

[
0.98^5 \approx 0.9039
]

## **Final Answer:**

**Probability that all 5 bulbs are non-defective ≈ 0.9039 (or 90.39%)**

## **Question 10:**

**Differentiate between discrete and continuous random variables with examples.**

## **Answer: Difference Between Discrete and Continuous Random Variables**

| **Discrete Random Variable**                | **Continuous Random Variable**                                             |
| ------------------------------------------- | -------------------------------------------------------------------------- |
| Takes **countable** values (whole numbers). | Takes **uncountable/infinite** values within a range.                      |
| Values are **separate and distinct**.       | Values form a **continuum**—any value in an interval.                      |
| Can be listed or counted.                   | Cannot be counted; measured on a scale.                                    |
| Probability of each value can be assigned.  | Probability of a specific value is **0**; only intervals have probability. |

## **Examples**

### **Discrete Random Variable Examples**

* Number of students in a class (e.g., 40, 41, 42…)
* Number of heads when flipping a coin 3 times
* Number of cars passing a signal in 1 minute
* Number of goals scored in a match

These are all **countable**.

### **Continuous Random Variable Examples**

* Height of students (e.g., 165.4 cm, 165.45 cm…)
* Weight of a person
* Time taken to complete a race
* Temperature of a room

These are **measured** and can take **any value** in a range.

## **Final Summary:**

* **Discrete = Countable values (whole numbers)**
* **Continuous = Measurable values (decimals allowed, infinite possibilities)**
